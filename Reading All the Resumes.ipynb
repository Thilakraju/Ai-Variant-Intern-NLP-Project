```python
import os
import re
import textract
import pandas as pd
import glob
import win32com.client
```

```python
# Converting All doc file into docx
word = win32com.client.Dispatch("Word.Application")

for i, doc in enumerate(glob.iglob("C:/Users/Sudeep/python files/DS Project 2/Resumes/Peoplesoft resumes/*.doc")):
    in_file = os.path.abspath(doc)
    wb = word.Documents.Open(in_file)
    out_file = re.sub(r'\.\w+$', '.docx', in_file)
    wb.SaveAs2(out_file, FileFormat=16) # file format for docx
    wb.Close()
word.Quit()
```

```python
# Removing The All original doc Files
dir_name = "C:/Users/Sudeep/python files/DS Project 2/Resumes/Peoplesoft resumes/"
paths = os.listdir(dir_name)

for item in paths:
    if item.endswith(".doc"):
        os.remove(os.path.join(dir_name, item))
```

```python
# Importing The all resumes of peoplesoft
file_path1 = [] 
category1  = []
directory1 = "C:/Users/Sudeep/python files/DS Project 2/Resumes/Peoplesoft resumes" 
for i in os.listdir(directory1):
    if i.endswith('.docx'):
        os.path.join(directory1, i)
        file_path1.append((textract.process(os.path.join(directory1, i))).decode('utf-8'))
        category1.append('PeopleSoft')
data1 = pd.DataFrame(data = file_path1 , columns = ['Resume_Text'])
data1['Category1'] = category1
data1
```
